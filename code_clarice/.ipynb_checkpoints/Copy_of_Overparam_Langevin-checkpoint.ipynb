```python
import numpy as np

```

```python
x0 = -1
A = np.random.randn(1)*2
y = A*x0
lam = np.abs(A*y)*1.1

niter = 50000 #number of iterations
M = 10000 # number of particles
tau = 0.001 #stepsize
```

Proximal langevin

```python
soft = lambda x,tau: np.sign(x)*np.maximum(np.abs(x)-tau,0)
x_ista = np.random.randn(M,)
for i in range(niter):
  g = A*(A*x_ista-y)
  x_ista = soft( x_ista - tau* g + np.random.randn(M,)*np.sqrt(2*tau), tau*lam )

```

u,v parameterization (symmetric)

```python

ru = lambda x: x[:M]
rv = lambda x: x[M:]
grad = lambda x,g, nm: np.concatenate((rv(x)*g - ru(x)/nm,ru(x)*g - rv(x)/nm ))
Grd = lambda x: lam*x  + grad( x, A*(A*ru(x)*rv(x)-y), (ru(x)**2+ rv(x)**2 ) )


#Corr = lambda x,nm: np.concatenate( (ru(x)/nm, rv(x)/nm) )


x = np.random.randn(M*2,)
for i in range(niter):
  x = x - tau*Grd(x) + np.random.randn(M*2,)*np.sqrt(2*tau)

u = ru(x)
v = rv(x)
x_hadamard_symm = u*v
```

u,v parametrization (asymmetric)


```python

ru = lambda x: x[:M]
rv = lambda x: x[M:]
grad = lambda x,g: np.concatenate((rv(x)*g+1.5/ru(x),ru(x)*g))
Grd = lambda x: lam*x  + grad( x, A*(A*ru(x)*rv(x)-y) )

#x = np.random.randn(M*2,)
#x[:M] = np.abs(x[:M])
for i in range(niter):
    #x[:M] = np.sqrt(x[:M]**2+tau*2)
    x = x - tau*Grd(x) + np.random.randn(M*2,)*np.sqrt(2*tau)
    
u = ru(x)
v = rv(x)
x_hadamard_asymm_v2 = u*v


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
ax = sns.kdeplot(x_hadamard_symm,color='blue', label='symm')
ax = sns.kdeplot(x_hadamard_asymm_v2,color='k', label='asymm - tony')

ax = sns.kdeplot(x_ista,color='green', label='ista')


#you might need to adjust the plot range for the 'true' distr
t= np.linspace(-3,.5,10000) #plot range


f = lambda x: np.abs(A*x - y)**2/2 +lam*np.abs(x);
z = np.exp(-f(t))
z = z/(np.sum(z)*(t[2]-t[1]))
plt.plot(t,z, 'c--',label='true')
#plt.xlim([-1,3])
plt.legend()
```
