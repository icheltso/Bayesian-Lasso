```python
import numpy as np
from solvers import *
import matplotlib.pyplot as plt
# import importlib
# importlib.reload(solvers)
```

```python
p = 1
m = 1
x0 = np.random.randn(p,)
x0[np.random.permutation(p)[:3*p//4] ]=np.zeros(3*p//4,)
A = np.random.randn(m,p)
y = A@x0
lam = np.max(np.abs(A.T@y))*.1

xinit = 7*np.ones(p,)
vinit = 2*np.ones(p,)

fval = lambda x: np.linalg.norm(A@x - y)**2 * 0.5
grad = lambda x: A.T@(A@x - y)
```

```python
from scipy import integrate

if p==1:
    def f(x):
        return np.exp(-fval(np.array([x]))-lam*np.abs(x))
    Total_mass = integrate.quad(f, -np.inf, np.inf)[0]
    print(Total_mass)
    density = lambda x: f(x)/Total_mass
    tgrid = np.linspace(-5,5,1000)
    density_grid = [density(t) for t in tgrid]
    plt.plot(tgrid,density_grid)
```

```python
def generate_samples_x(Iterate,init, n, burn_in=1000):
    x = init

    #burn in 
    for i in range(burn_in):
        x = Iterate(x)
    p = len(x)

    #record n samples
    samples = np.zeros((n,p))
    for i in range(n):
        x = Iterate(x)
        samples[i,:] = x.reshape(-1)
        
    return samples

        
```

```python
tau = 0.001 #stepsize
gamma = tau*.1

Iterate = lambda x: prox_langevin(x, grad, tau, gamma, lam)
n = 100000
burn_in = 10000
samples_proxl1 = generate_samples_x(Iterate, xinit, n, burn_in)
```

```python
import matplotlib.pyplot as plt
plt.hist(samples_proxl1,100,density=True)

if p==1:
    plt.plot(tgrid,density_grid)
plt.show()
```

```python
tau = 0.01 #stepsize
print(lam)
Iterate = lambda x: one_step_uv(x, grad, tau, lam)

uvinit = np.concatenate((xinit,vinit))
print(uvinit.shape)
burn_in = 1000
n = 100000
samples_uv = generate_samples_x(Iterate, uvinit, n, burn_in)
```

```python
plt.hist(samples_uv[:,:p]*samples_uv[:,p:],100,density=True)
if p==1:
    plt.plot(tgrid,density_grid)
plt.show()
```

```python
tau = 0.01 #stepsize

Iterate = lambda x: one_step_MALA_uv(x, fval, grad, tau, lam)

uvinit = np.concatenate((xinit,vinit))
burn_in = 2000
n = 100000
samples_uv_mala = generate_samples_x(Iterate, uvinit, n, burn_in)
```

```python
plt.hist(samples_uv_mala[:,:p]*samples_uv_mala[:,p:],100,density=True)
if p==1:
    plt.plot(tgrid,density_grid)
plt.show()
```

```python
Iterate = lambda x_eta: one_step_gibbs(x_eta, A,y, lam)
x_eta = np.ones(2*p,)
burn_in = 1000
n = 100000
gibbs_samples = generate_samples_x(Iterate, x_eta, n, burn_in)
```

```python
plt.hist(gibbs_samples[:,:p],100,density=True)
if p==1:
    plt.plot(tgrid,density_grid)
plt.show()
```

#checking the effective sample dimensions

```python
#effective sample size
import numpy as np
import arviz as az
import matplotlib.pyplot as plt
```

```python

# Compute Effective Sample Size (ESS)

samples1 = gibbs_samples[:,:p]
samples1 = samples_uv[:,:p]*samples_uv[:,p:]
samples2 = samples_uv_mala[:,:p]*samples_uv_mala[:,p:]
# samples2 = samples_proxl1


idata_algo1 = az.convert_to_inference_data(np.expand_dims(samples1, 0))
idata_algo2 = az.convert_to_inference_data(np.expand_dims(samples2, 0))


ess_algo1 = az.ess(idata_algo1)
ess_algo2 = az.ess(idata_algo2)
```

```python

# Print the results
print("Effective Sample Size (ESS) for Algorithm 1:")
print(ess_algo1)

print("\nEffective Sample Size (ESS) for Algorithm 2:")
print(ess_algo2)
```

```python
# Visualization
labels = []
for i in range(p):
    labels.append('Dimension '+ str(i))
# labels = ['Dimension 1', 'Dimension 2']
ess_values_algo1 = ess_algo1['x'].values
ess_values_algo2 = ess_algo2['x'].values

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, ess_values_algo1, width, label='Algorithm 1')
rects2 = ax.bar(x + width/2, ess_values_algo2, width, label='Algorithm 2')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_ylabel('ESS')
ax.set_title('Effective Sample Size by Dimension and Algorithm')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()
plt.show()
```
